## Create corpus

```python
corpus = []
for i in range(len(data)):
  corpus.append(data[i]['question'])
```

```python
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel

# Download PhoBERT Tokenizer and Model
tokenizer = AutoTokenizer.from_pretrained('vinai/phobert-base')
embeder = AutoModel.from_pretrained('vinai/phobert-base')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
if device.type == 'cpu':
    raise RuntimeError("No GPU found")

embeder.to(device)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')

corpus_embeddings = model.encode(corpus, convert_to_tensor=True)
```

## Semantic search

```python
from sentence_transformers.cross_encoder import CrossEncoder

model_cross_encoder = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
```

```python
from sentence_transformers import util
import numpy as np
def semantic_search(query, topk=10):
  query_embedding = model.encode(query, convert_to_tensor=True)
  hits = util.semantic_search(query_embedding, corpus_embeddings, top_k=topk)[0]

  print("Bi-encoder model")
  print("Query:", query)
  for hit in hits:
    print(f"{hit['score']:.4f}\t{data[hit['corpus_id']]['question']}")

  # re-ranks
  cross_inp = [[query, data[hit['corpus_id']]['question']] for hit in hits]
  cross_scores = model_cross_encoder.predict(cross_inp)
  ranks = np.argsort(cross_scores)[::-1][:5]

  # Sort results by the cross-encoder scores
  print()
  print("Re-ranks by cross-encoder")
  print(f"Query: {query}")
  for rank in ranks:
    print(f"{hits[rank]['score']:.4f}\t{data[hits[rank]['corpus_id']]['question']} - {data[hits[rank]['corpus_id']]['answers']}")
```

```python
result_bi_encoder = semantic_search("nộp thuế thì đóng cho ai?")
```
